# Beta - VAE

```python
df_1 = pd.read_csv('checkpoints/log_beta_VAE_old.csv')
df_2 = pd.read_csv('checkpoints/log_beta_VAE.csv')

df = pd.concat([df_1, df_2], ignore_index=True)
del df_1, df_2
```

```python
df
```

```python
#grafico academico via seaborn
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 4))
sns.lineplot(data=df, x='step', y='loss', color='darkblue')
# sns.lineplot(data=df, x='step', y='val_loss', label='Validation Loss', color='orange')
plt.title('Loss over Steps for Beta VAE')
plt.xlabel('Step')
plt.ylabel('Loss')
plt.legend()
plt.show()
# df.plot(x='step', y='loss', label='Loss', figsize=(10, 5))
```
